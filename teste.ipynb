```python
import pandas as pd
import streamlit as st
from sklearn.linear_model import LinearRegression
```

Incluindo dataframe com os dados e primeira checagem com o metodo "head()"

```python
df = pd.read_csv('./pizza_data.csv')
df.head()
```

Pré-processamento dos dados

```python
df['Size'].unique()
```

```python
df = df.loc[~df['Size'].isin(['Medium', 'Large', 'Personal', 'Small', 'Mini', 'Jumbo', 'X Large (16")'])]
```

```python
df['Size'].unique()
```

```python
df['Price'].unique()
```

```python
df = df.drop(['Company', 'Pizza Name', 'Type'], axis=1)
df.head()
```

```python
df = df.replace({'Small (10")': 10, 'Medium (12")': 12, 'Large (14")': 14, 'X-Large (16")': 16})
df['Price'] = df['Price'].str.replace('$', '', regex=False)
df.head()
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df.head()
```

```python
df['Price'] = df['Price'] * 6.11
df.head()
```

```python
df.to_csv('dataframe_tratado.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.use('TkAgg') 

df.plot(kind="scatter", x="Size", y="Price")
plt.show()
```

```python
modelo = LinearRegression()

x = df[['Size']]
y = df[['Price']]

modelo.fit(x,y)
```
